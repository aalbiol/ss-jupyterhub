```python
import pandas as pd
import glob
import os
from pathlib import Path
```

```python
homedir=os.getenv("HOME")
notasdir=os.path.join(homedir,'notas')
notas_limpias_dir=os.path.join(homedir,'notas_limpias')
print(notasdir)
Path.mkdir(notas_limpias_dir,parents=True,exist_ok=True)
```

```python
def procesa_csv(csv_filename,notas_limpias_dir):
    df=pd.read_csv(csv_filename)
    df2=df[ df['email'].notna()]
    df2.index=df2.email
    nameout=os.path.join(notas_limpias_dir, os.path.basename(csv_filename))
    df3=df2[['score','max_score']]
    df3.to_csv(nameout,header=False)
    return df3 
    
```

```python
ficheros =glob.glob(notasdir+"/*.csv")
print(len(ficheros), " ficheros")
```

```python
for f in ficheros:
    a=procesa_csv(f,notas_limpias_dir)
```
