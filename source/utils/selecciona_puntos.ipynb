```python
import matplotlib.pylab as plt
import numpy as np

```

```python
t=np.linspace(0,4*np.pi)+0.2
x=np.sin(t)+ np.random.randn(t.shape[0])*0.3
```

```python
%matplotlib widget
f,a = plt.subplots()

pos = []
def onclick(event):
    print("clicado")
    pos.append((event.xdata,event.ydata))
    print(pos)
    a.plot(pos[-1][0], pos[-1][1],'*r')
    a.set_title(f'Numero de clicks = {len(pos)}')
    
a.plot(x) # grafica anterior


    
f.canvas.mpl_connect('button_press_event', onclick)
```

```python
plt.close(f)
```
