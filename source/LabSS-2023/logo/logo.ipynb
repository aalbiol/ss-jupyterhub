```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python

e1=np.zeros(500)
e2=np.ones(1000)

envolvente=np.concatenate((e1,e2,0.5*e2,e2,e1))

B=1/70
A=[1, -(1-B)]

ef=signal.lfilter(B,A,envolvente)
```

```python
_=plt.plot(envolvente)
_=plt.plot(ef)
```

```python
f1=1/100
f2=1/50
f3=1/75
frec1=np.ones(1500)*f1
frec2=np.ones(1000)*f2
frec3=np.ones(1500)*f3
frec=np.concatenate((frec1,frec2,frec3))
fase=2*np.pi*np.cumsum(frec)
```

```python
s=ef*np.cos(fase)
```

```python
_=plt.plot(s)
_=plt.axis('off')
```
