```python
def procesa_csv(csv_filename,notas_limpias_dir):

    nombre_columna=os.path.basename(csv_filename).replace(".csv","")
    print(nombre_columna)
    df=pd.read_csv(csv_filename)
    df2=df[ df['email'].notna()]
    df2.index=df2.email
    nameout=os.path.join(notas_limpias_dir, os.path.basename(csv_filename))
    df3=df2[['score','max_score','email']]
    df3['DNI']=""
    for index, row in df.iterrows():
        if row['email'] in tabla:
            row['DNI']=tabla[row['email']]
        else:
            print("NO en la lista:", row['email'])
    #print(">>>>>>>>>>>>>>>>>>")
    #print(df3['email'])
    #print(">>>>>>>>>>>>>>>>>>")
    #df3=df3[df3['email'].str.contains('upv.es', case=False, na=False)]
    print(len(df3))
    #df3['DNI']=dflista[df3['email']]
    #df3['score']=df3['score'].fillna(0)
    df3['score'].fillna(0)
    df3[nombre_columna]=10*df3['score']/df3['max_score']
    df3[nombre_columna] = df3[nombre_columna].round(0).astype('int')
    df4=df3[['DNI',nombre_columna]]
    df4.to_csv(nameout,header=True,index=False,sep=';')
    df3.to_csv("kk.csv",header=True,index=False,sep=';')
    return df4 

def procesa_csv2(csv_filename,notas_limpias_dir):

    nombre_columna=os.path.basename(csv_filename).replace(".csv","")
    print(nombre_columna)
    df=pd.read_csv(csv_filename)
    df2=df[ df['email'].notna()]
    df2.index=df2.email
    nameout=os.path.join(notas_limpias_dir, os.path.basename(csv_filename))
    df3=df2[['score','max_score','email']]
    df3['score'].fillna(0)
    
    alumnos=[]
    for index, row in df3.iterrows():
        alumno={}
        if row['email'] in tabla:
#            if "allovil2" in row['email']:
#                print("allovil2 found")
            alumno['DNI']=tabla[row['email']]
            alumno[nombre_columna]=int(round(row['score']/(0.0001+row['max_score'])*10))
            alumnos.append(alumno)
        else:
            print("NO en la lista:", row['email'])

    #print(alumnos)
   
    df4=pd.DataFrame(alumnos)
    #print(df4)
    df4.to_csv(nameout,header=True,index=False,sep=';')
    
    return df4 
    
```
